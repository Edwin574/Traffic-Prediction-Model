```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
traffic = pd.read_csv("DataSet/train13519.csv")
traffic.tail(20)
```

```python
traffic[['date', 'time']] = traffic['timestamp'].str.split(' ', 1, expand=True)
traffic_new = traffic.drop("timestamp", 1)

traffic_new
```

```python
traffic_new["date"].min(), traffic_new["date"].max()
```

```python
# Changing the time interval to be hourly and also aggregating the traffic so as to shorten the dataset and get a better chance to predict traffic flow

traffic_new['time'] = pd.to_datetime(traffic_new['time'])

traffic_hourly = traffic_new.resample("H", on="time").hourly_traffic_count.sum()

traffic_hourly.head(20)
```

```python
traffic_hourly = pd.DataFrame(traffic_hourly)
traffic_hourly
```

Conclusion: This is a data of traffic in Road A for a period of 3 months.

```python
# Split the dataset to training and testing data 
Y_train_df = traffic_hourly[:672] 
Y_test_df = traffic_hourly[673:1344]
```

```python
# Import the models we will use to predict
from statsforecast import StatsForecast
from statsforecast.models import AutoARIMA, ETS
```

```python
season_length = 24
horizon = len(Y_test_df)

models = [AutoARIMA(season_length=season_length), ETS(season_length=season_length, model='ZMZ')]

model = StatsForecast(df=Y_train_df, models=models, freq='H', n_jobs=-1)
```
